"""
Q.6 Write a  program which accepts a strings and count the frequency of each vowel.
"""

```python
st = input("Enter a string: ").lower()

vowels = "aeiou"
freq = {"a":0,"e":0,"i":0,"o":0,"u":0}   # dictionary to store count

for ch in st:
    if ch in vowels:
        freq[ch] += 1

print("Vowel frequencies:")
for v, c in freq.items():
    print(v, ":", c)
```

"""

Eg: Robert Brett Roser output :R.B.Roser.
"""

```python
st = input("Enter a string: ")
parts = st.split()
last = parts[-1]
abbs = ""
for name in parts[:-1]:
    abbs = abbs + name[0].upper() +"."
result = abbs + last
print(result)
```

"""
Q.13 Write a Python program to add 'ing' at the end of a given string (length should be at least 3).

If the string length of the given string is less than 3, leave it unchanged.
Sample String : 'abc'
Ex pected Result : 'abcing'

"""

```python
st = input("Enter a string: ")

if len(st) >= 3 and st[-3:] != "ing":
    print(st+"ing")
elif len(st) >= 3 and st[-3:] == "ing":
    print(st+"ly")
else:
    print(st)
```

```python
"""Write a program to print those string that is pallindrome.
sample list= [‘level’, ‘hello’, ‘1221’, ‘789, ‘wow’]
sample output=[‘level’, ‘1221’ ‘wow’]
"""
l = ["level", "hello", 1221, 789, "wow"]
u = []
for i in l :
    i = str(i)
    if i == i[::-1]:
        u.append(i)
print(u)
```
